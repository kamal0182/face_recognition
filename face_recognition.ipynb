```python
from retinaface import RetinaFace
import cv2
import matplotlib.pyplot as plt
image_path = "/home/younes/projects/face_recognition/20260216_190013.jpg"
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
faces = RetinaFace.detect_faces(image_path)
```

```python
face_key = list(faces.keys())[1]
bbox = faces[face_key]["facial_area"] 
x1, y1, x2, y2 = bbox
face_crop = img_rgb[y1:y2, x1:x2]
plt.imshow(face_crop)
plt.axis("off")
plt.show()
```

```python
cv2.imwrite("carte_image.jpg", cv2.cvtColor(face_crop, cv2.COLOR_RGB2BGR))
```

```python
from deepface import DeepFace

embedding = DeepFace.represent(face_crop, model_name="Facenet")
print(embedding)
```
